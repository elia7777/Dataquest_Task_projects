# Визуализация доходов, основанных на специальностях колледжа

Мы будем работать с dataset о результатах работы студентов, окончивших колледж в период между 2010 и 2012 годами.
Dataset выпустили его на своем репозитории [Github.](https://github.com/fivethirtyeight/data/tree/master/college-majors)

Каждая строка в dataset представляет отдельную специализацию в колледже и содержит информацию о показателях занятости, средней заработной плате и многом другом. 

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.head()
```

```python
recent_grads.tail(3)
```

```python
recent_grads.info()
```

Удалим нулевые значения

```python
recent_grads = recent_grads.dropna()
```

```python
recent_grads.info()
```

## Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
```

Наблюдения:
    - Студенты в более популярных специальностях не зарабатывают больше денег
    - Чем менее популярная специальность тем выше уровень безработицы до определенного момента
    - Чем больше доля женщин в специальности тем ниже медиана зарплаты

## Pandas, Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r-1]].plot(kind='hist', rot=20)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=20)
```

```python
recent_grads['Median'].hist(bins=18, range=(22000, 110000))
```

Самая популярная зарпалат лежит в диапозоне от 30 000\$ до 40 000\$
